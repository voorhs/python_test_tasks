## All Valid Inputs

```python
from tqdm.contrib.concurrent import process_map
import itertools as it
import random


def solve(desk_size):
    m, n = desk_size
    return m, n, solution.main([m, n])

# according to given task
desk_sizes = list(it.product(range(4, 101), range(4, 101)))

# to make progress bar more stable
random.shuffle(desk_sizes)

# solve in parallel
res = process_map(solve, desk_sizes, total=97*97, chunksize=1)
```

```python
errors = [x for x in res if x[2] is None]
len(errors)
```

## Animation

```python
m, n = 15, 20
points = np.array(solution.main([m, n]))
print(len(points), *points)
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as anm
from IPython import display

# alternate position for coloring
board = (np.add.outer(range(n), range(m)) + 1) % 2

fig, ax = plt.subplots(1, 1)
ax.imshow(board, cmap='Pastel1', interpolation='nearest', alpha=1)
ax.axis('off')
ax.invert_yaxis()

def animate(i):
    pnt = points[i:i+2]
    # ax.clear()
    ax.plot(pnt[:, 0], pnt[:, 1], lw=1)
    

anim = anm.FuncAnimation(fig, animate, frames=len(points)-1, interval=50, repeat=False)

# converting to an html5 video
video = anim.to_html5_video()
  
# embedding for the video
html = display.HTML(video)
  
# draw the animation
display.display(html)

plt.close()
```
